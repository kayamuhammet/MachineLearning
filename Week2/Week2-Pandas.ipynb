### Indexing and Slicing

```python
print(df.loc[:,"age"]) # sutun getirmenin bir diğer yolu
```

```python
print(df.loc[:2,"age"]) #pandas'da yazılan sayı dahildir.
```

```python
print(df.loc[:2, "name":"height"]) # name'den height'e kadar olan kolonların ilk 2 satırını yazdı. 2 dahil
```

```python
print(df.loc[:2, ["name","height"]]) #sadece name ve height sutunları alındı
```

```python
print(df.loc[::-1]) # indexleri tersden yazdırdı.
```

```python
print(df.loc[:,:"age"]) # age'e kadar olan tum satırları yazdırdı. age dahil.
```

```python
print(df.iloc[:,[1]]) # 1.indexdeki kolonun tum satırlarını verir.
```

```python
# iloc vs loc : iloc sadece int değerleri ile arama yapabiliyor. loc'da string de var.
```

## Filtering Pandas Data Frame

```python
filtre = df.age < 20  # yası 20den kucuk olanları filtreledik.
df["bool"] = filtre
print(df.loc[:,["age","bool"]])
```

```python
type(filtre)
```

```python
# numpy'da -> vektör, pandas'da -> series deniliyor. Serilerin birleşmesine ise dataframe denir.
```

```python
filtre_data = df[filtre] # sadece kosulu sağlayanlar yazıldı.
print(filtre_data)
```

```python
filtre2 = df.height < 160
filtre_data2 = df[filtre & filtre2]
print(filtre_data2)
```

## List Comprehension

```python
import numpy as np
avg = df.age.mean() # pandas ile ortalama bulma
print(avg)

avg_np = np.mean(df.age) # numpy ile ortalama bulma
print(avg_np)
```

```python
print(df)
```

```python
df.dropna(inplace=True) # dropna ile Nan'ları sildik. inplace=True, olayı kalıcı yapar:
print(df)
```

```python
print(df.age.mean()) # ortalama yas alındı.
df["ortalama"] = ["Ortalamanın altında" if df.age.mean() > each else "ortalamanın üzerinde" for each in df.age]
df
```

```python
# list comprehension ile tek bir satırda if-else ve for kullandık.
```

```python
df.columns = [each.upper() for each in df.columns] # kolonları büyülttük.
df.columns
```

## Concatenating Data

```python
df
```

```python
df.drop(["ORTALAMA"],axis = 1) # axis'in yanına inplace=True eklersek kalıcı siler.  axis = 1 sutun , axis = 0 satırı siler.
df
```

```python
df
```

```python
#verticle(dikey) birleştirme:->  #concat() metodu ile iki dataframe'i birleştirdik  axis = 0
data1 = df.head()     
data2 = df.tail()
data_concat = pd.concat([data1,data2],axis=0)
data_concat
```

```python
#horizontal(yatay) birleştirme
data_contact2 = pd.concat([data1,data2],axis=1) #axis= 1 yazarak stunları birleştirdik
data_contact2
```
